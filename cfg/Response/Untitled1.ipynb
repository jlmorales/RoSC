```python
import pandas as pd 

def readAndClean(filepath, col_subset):
    filepath = filepath
    data = pd.read_excel(filepath)
    df = pd.DataFrame(data)
    data = df.dropna(subset=[col_subset])
    data = data[:-1]
    dataframe = data
    dataframe = dataframe.dropna(subset=[col_subset])
    return dataframe

def processData(dataframe, col_subset, day):
    dataframe[col_subset] = pd.to_datetime(dataframe[col_subset])
    dataframe['by_hour'] = dataframe[col_subset].dt.floor('h')
    dataframe['by_day'] = dataframe[col_subset].dt.floor('d')
    newdf = dataframe.loc[dataframe['by_day'] == day]
    byhr = newdf.groupby(['by_hour']).count()
    return byhr


filepath = "C4G Call Center Call Detail .xlsx"
col_subset = 'Call Start Time'
df = readAndClean(filepath,col_subset)
df = processData(df,col_subset, '2019-02-02')




```
